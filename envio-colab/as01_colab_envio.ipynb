### 1. Normalização
Realizar as seguintes ações de normalização no arquivo de entrada e gerar o arquivo de saída Shakespeare_Normalized.txt.

1. Lower case reduction  
2. Accent and diacritic removal  
3. Canonicalizing of acronyms, currency, date and hyphenated words  
4. Punctuation removal (except currency and date).  
5. Special characters removal  

````python
# imports
import re
import unicodedata


# (1.1) Iterando o dicionario por key e value, aplicando o toLowerCase em todas as linhas do dicionário já mapeadas

for linha, valor in shakespeare_dict.items():

  ## (1.1) Colocando cada linha com lowerCase
  texto_normalizado = valor.lower()

  ## (1.2) Removendo acentos e diatrícos
  texto_normalizado = unicodedata.normalize('NFD', texto_normalizado)
  texto_normalizado = re.sub(r'[\u0300-\u036f]', '', texto_normalizado)

  ## (1.3) Canonização de siglas, moeda, data e palavras hifenizadas

  # Juntando string com hifem
  texto_canonizado =  texto_normalizado.replace("-", "")
  # Tratando data para manter DD/MM/YYYY
  texto_canonizado = re.sub(r'(\d{2})/(\d{2})/(\d{4})', r'\3-\2-\1', texto_canonizado)

  ## (1.4) Removendo pontuação exceto caso sej data.

  # Remove todas as pontuações
  texto_sem_pontuacao = re.sub(r'[^\w\s]|_', '', texto_canonizado)

  # Faz uma restauração caso encontre o padrão data seja encontado (DD/MM/YYYY)
  texto_sem_pontuacao = re.sub(r'\b\d{2}/\d{2}/\d{4}\b', lambda match: match.group(), texto_sem_pontuacao)


  ## (1.5) Removendo caracteres especiais
  texto_sem_caracteres_especias = re.sub(r'[^\w\s]', '', texto_sem_pontuacao)

  # Salvando valor na linha novamente do dicionario
  shakespeare_dict[linha] = texto_sem_caracteres_especias


````

````python
dict_white_space_tokenization = {}
dict_word_tokenizer = {}
dict_tree_bank_tokenizer = {}
dict_word_punct_tokenizer = {}
dict_tweet_tokenizer = {}
dict_mwe_tokenizer = {}
dict_textblob_tokenizer = {}
dict_gensim_tokenizer = {}

for linha, valor in shakespeare_dict_normalize.items():

  # 1. White Space Tokenization
  dict_white_space_tokenization[linha] = valor.split()

  # 2. NLTK: Word Tokenizer
  dict_word_tokenizer[linha] = word_tokenize(valor)

  # 3. NLTK: Treebank Tokenizer
  tokenizer = TreebankWordTokenizer()
  dict_tree_bank_tokenizer[linha] = tokenizer.tokenize(valor)

  # 4. NLTK: Word Punctuation Tokenizer
  tokenizer_punct = WordPunctTokenizer()
  dict_word_punct_tokenizer[linha] = tokenizer_punct.tokenize(valor)

  # 5. NLTK: Twwet Tokenizer
  tknzr = TweetTokenizer()
  dict_tweet_tokenizer[linha] = tknzr.tokenize(valor)

  # 6. NLTK: MWE Tokenizer
  tknzr_mwe = MWETokenizer([('poor', 'contempt'), ('good', 'chamber')])
  dict_mwe_tokenizer[linha] = tknzr_mwe.tokenize(valor.split())

  # 7. TextBlob
  blob = TextBlob(valor)
  dict_textblob_tokenizer[linha] = blob.words

  # 9. Gensim Word Tokenizer
  dict_gensim_tokenizer[linha] = Dictionary([word_tokenize(valor.lower())])




###############################
#          SALVAMENTO         #
###############################

# Salvando em arquivo 1 White Space Tokenization
with open("saidas/AS01/shakespeare_Normalized_Tokenized01.txt", 'w') as arquivo:
  for value in dict_white_space_tokenization.values():
    arquivo.write(str(value) + '\n')

# Salvando em arquivo 2 NLTK: Word Tokenizen
with open("saidas/AS01/shakespeare_Normalized_Tokenized02.txt", 'w') as arquivo:
  for value in dict_word_tokenizer.values():
    arquivo.write(str(value) + '\n')

# Salvando em arquivo 3 NLTK: TreeBank Tokenizer
with open("saidas/AS01/shakespeare_Normalized_Tokenized03.txt", 'w') as arquivo:
  for value in dict_tree_bank_tokenizer.values():
    arquivo.write(str(value) + '\n')

# Salvando em arquivo 4 NLTK: Word Punctuation Tokenizer
with open("saidas/AS01/shakespeare_Normalized_Tokenized04.txt", 'w') as arquivo:
  for value in dict_word_punct_tokenizer.values():
    arquivo.write(str(value) + '\n')

# Salvando em arquivo 5 NLTK: Tweet Tokenizer
with open("saidas/AS01/shakespeare_Normalized_Tokenized05.txt", 'w') as arquivo:
  for value in dict_tweet_tokenizer.values():
    arquivo.write(str(value) + '\n')

# Salvando em arquivo 6 NLTK: MWE Tokenizer
with open("saidas/AS01/shakespeare_Normalized_Tokenized06.txt", 'w') as arquivo:
  for value in dict_mwe_tokenizer.values():
    arquivo.write(str(value) + '\n')

# Salvando em arquivo 7 TextBlob
with open("saidas/AS01/shakespeare_Normalized_Tokenized07.txt", 'w') as arquivo:
  for value in dict_textblob_tokenizer.values():
    arquivo.write(str(value) + '\n')

# Salvando em arquivo 9 Gensim Dict
with open("saidas/AS01/shakespeare_Normalized_Tokenized09.txt", 'w') as arquivo:
  for value in dict_gensim_tokenizer.values():
    arquivo.write(str(value) + '\n')
````

Tratando o Lematizador, Potter e Snow e Gerando Csv


1. Token
2. Número e Ocorrência do token no documento resultante
3. Tamanho em caracteres de cada token do vocabulário

````python

#################
# Lematizador   #
#################
list_occurrence_number_lemmatized = []
list_word_length_lemmatized = []

#################
#   Porter      #
#################
list_occurrence_number_porter = []
list_word_length_porter = []

#################
#   Snowball    #
#################
list_occurrence_number_snowball = []
list_word_length_snowball = []

# Removendo a repetição de tokens caindo de 9517 tokens para 3102
list_lemmatzed_unique_norepeat = list(set(list_lemmatzed_unique))
list_porter_unique_norepeat = list(set(list_porter_unique))
list_snowball_unique_norepeat = list(set(list_snowball_unique))

for each_word in list_lemmatzed_unique_norepeat:
  list_occurrence_number_lemmatized.append(list_lemmatzed_unique.count(each_word))
  list_word_length_lemmatized.append(len(each_word))

for each_word in list_porter_unique_norepeat:
  list_occurrence_number_porter.append(list_porter_unique.count(each_word))
  list_word_length_porter.append(len(each_word))

for each_word in list_snowball_unique_norepeat:
  list_occurrence_number_snowball.append(list_snowball_unique.count(each_word))
  list_word_length_snowball.append(len(each_word))



dict_lemmatized_csv = {

  "Token": list_lemmatzed_unique_norepeat,
  "N_Ocorrencia": list_occurrence_number_lemmatized,
  "Length": list_word_length_lemmatized

}

dict_porter_csv = {"Token": list_porter_unique_norepeat,
  "N_Ocorrencia": list_occurrence_number_porter,
  "Length": list_word_length_porter}


dict_snowball_csv = {"Token": list_snowball_unique_norepeat,
  "N_Ocorrencia": list_occurrence_number_snowball,
  "Length": list_word_length_snowball
}

df1 = pd.DataFrame(dict_lemmatized_csv)
df1.to_csv("saidas/AS01/Shakespeare_Vocabulary_Lemmatized.csv")

df2 = pd.DataFrame(dict_porter_csv)
df2.to_csv("saidas/AS01/Shakespeare_Vocabulary_Porter.csv")

df3 = pd.DataFrame(dict_snowball_csv)
df3.to_csv("saidas/AS01/Shakespeare_Vocabulary_Snowball.csv")
````
