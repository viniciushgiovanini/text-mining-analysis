```python
# Realizando Leitura do Arquivo, e mapeando cada linha para um dicionário, conseguindo assim controlar cada alteração por linha.
# Removendo as quebras de linhas e pulando linhas que são vazias

shakespeare_dict = {}

cont = 1

with open("../datasets/Shakespeare.txt", 'r') as arquivo:
  
  
  for linha in arquivo:
    shakespeare_dict[f'Linha {cont}'] = linha.replace("\n", "")
    cont+=1
```

### 1. Normalização
Realizar as seguintes ações de normalização no arquivo de entrada e gerar o arquivo de saída Shakespeare_Normalized.txt.

1. Lower case reduction  
2. Accent and diacritic removal  
3. Canonicalizing of acronyms, currency, date and hyphenated words  
4. Punctuation removal (except currency and date).  
5. Special characters removal  

```python
# imports
import re
import unicodedata


# (1.1) Iterando o dicionario por key e value, aplicando o toLowerCase em todas as linhas do dicionário já mapeadas

for linha, valor in shakespeare_dict.items():
  
  ## (1.1) Colocando cada linha com lowerCase
  texto_normalizado = valor.lower()
  
  ## (1.2) Removendo acentos e diatrícos
  texto_normalizado = unicodedata.normalize('NFD', texto_normalizado)
  texto_normalizado = re.sub(r'[\u0300-\u036f]', '', texto_normalizado)
  
  ## (1.3) Canonização de siglas, moeda, data e palavras hifenizadas

  # Juntando string com hifem
  texto_canonizado =  texto_normalizado.replace("-", "")
  # Tratando data para manter DD/MM/YYYY
  texto_canonizado = re.sub(r'(\d{2})/(\d{2})/(\d{4})', r'\3-\2-\1', texto_canonizado)
  
  ## (1.4) Removendo pontuação exceto caso sej data.
  
  # Remove todas as pontuações
  texto_sem_pontuacao = re.sub(r'[^\w\s]|_', '', texto_canonizado)
  
  # Faz uma restauração caso encontre o padrão data seja encontado (DD/MM/YYYY)
  texto_sem_pontuacao = re.sub(r'\b\d{2}/\d{2}/\d{4}\b', lambda match: match.group(), texto_sem_pontuacao)
  
  
  ## (1.5) Removendo caracteres especiais  
  texto_sem_caracteres_especias = re.sub(r'[^\w\s]', '', texto_sem_pontuacao)
  
  # Salvando valor na linha novamente do dicionario
  shakespeare_dict[linha] = texto_sem_caracteres_especias
  
  
  
```
